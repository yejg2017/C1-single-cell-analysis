### Combine data and save

```r
#DataSet=c()
GroupSet=c()

for(sub_data in sub_data_path){
    df=read.csv(sub_data,header = TRUE)[,-1]
    #DataSet<-rbind(DataSet,df)
    
    splited=str_split(basename(sub_data),"_",n=3)
    group=paste0(splited[[1]][1],"_",splited[[1]][2])
    GroupSet=c(GroupSet,rep(group,dim(df)[2]))
}
```

```r
GroupSet=str_replace_all(GroupSet,pattern = "P11-2_H",replacement = "P11-1_H")
GroupSet=str_replace_all(GroupSet,pattern = "P11-2_L",replacement = "P11-1_L")
GroupSet=str_replace_all(GroupSet,pattern = "P11-2_M",replacement = "P11-1_M")
#GroupSet=GroupSet[!str_detect(GroupSet,pattern = "P11-N_*")]
#GroupSet=GroupSet[!str_detect(GroupSet,"P2-N_*")]
P2_N=GroupSet[str_detect(GroupSet,"P2-N_*")]
P11_N=GroupSet[str_detect(GroupSet,pattern = "P11-N_*")]
```

```r
geneCounts=geneCounts[-seq(dim(geneCounts)[1]-4,dim(geneCounts)[1]),]
geneCounts=na.omit(geneCounts)
```

#### 去除线粒体的RNA,去掉Mt, LOC, Gm开头的基因

```r
mt.genes=rownames(geneCounts)[str_detect(string = rownames(geneCounts),pattern = "^mt-")]
Gm.genes=rownames(geneCounts)[str_detect(string = rownames(geneCounts),pattern = "^Gm\\d+")]
loc.genes=rownames(geneCounts)[str_detect(string = rownames(geneCounts),pattern = "^Loc\\w+")]
keep.genes=setdiff(setdiff(setdiff(rownames(geneCounts),mt.genes),Gm.genes),loc.genes)
```

```r
v=get.variable.genes(geneCounts[keep.genes,],min.cv2 =2,do.plot = TRUE)
var.genes<-as.character(rownames(v))[v$p.adj<0.05]
```

```r
#obj=CreateSeuratObject(raw.data = geneCounts[-seq(dim(geneCounts)[1]-4,dim(geneCounts)[1]),!(GroupSet%in%P11_N | GoupSet%in%P2_N)],project = "SingleCell")
#obj@meta.data$group=as.factor(GroupSet[!(GroupSet%in%P11_N)])#roupSet%in%P2_N)])
obj=CreateSeuratObject(raw.data = geneCounts[keep.genes,!GroupSet%in%P2_N],project ="SingleCell",min.cells =6,min.genes = 500)
# obj@meta.data$group=as.factor(GroupSet[!(GroupSet%in%P11_N)])#|GroupSet%in%P2_N)])

#obj=CreateSeuratObject(raw.data = geneCounts[keep.genes,],project ="SingleCell",min.cells =200,min.genes =500)
#obj@meta.data$group=as.factor(GroupSet[!GroupSet%in%P2_N])
obj@meta.data$group=as.factor(GroupSet[colnames(geneCounts)%in%colnames(obj@raw.data)])
obj=NormalizeData(obj,display.progress =FALSE)
obj=ScaleData(obj,display.progress = FALSE)
obj=FindVariableGenes(obj,do.plot = TRUE,display.progress = FALSE)
```

####  Create dataset to GO analysis

```r
group.count<-function(object,group,func="median"){
    cell.names<-object@cell.names
    cell.s<-cell.names[object@meta.data$group%in%group]
    df<-object@raw.data[,cell.s]
    
    if(func=="median"){
       value<-as.integer(apply(df,1,median))
    }
    if(func=="scale"){
        value<-apply(df,1,sum)
        #value<-as.numeric(scale(value,center = TRUE,scale =TRUE))
        #value<-log10(1+value/sum(value)*10^4)
    }
    #x<-data.frame(value)
    #colnames(x)<-group
    return(value)
}

groups<-obj@meta.data$group
```

```r
result<-lapply(unique(groups),group.count,object=obj,func="scale")
n.obs <- sapply(result, length)
seq.max <- seq_len(max(n.obs))
mat<-data.frame((sapply(result, "[", i = seq.max)))
colnames(mat)<-unique(groups)
rownames(mat)<-rownames(obj@raw.data)
#write.table(mat,file ="./scale.csv",col.names = TRUE,sep = ",",row.names =TRUE)
```

```r
x=CreateSeuratObject(raw.data = mat)
x=NormalizeData(x,display.progress =FALSE)
x=ScaleData(x,display.progress = FALSE)
x=FindVariableGenes(x,do.plot = F)
```

```r
annlab=obj@ident
logcount=as.matrix(obj@raw.data[obj@var.genes,])
correlation_mat=cor(logcount)
```

```r
OrderLab=function(data,lab){
    n.unique=length(unique(lab))
    order.df=data.frame()
    order.lab=c()
    for(i in 1:n.unique){
        idx=which(lab==i)
        sub.df=data[idx,]
        sub.lab=lab[idx]
        order.df=rbind.data.frame(order.df,sub.df)
        order.lab=c(order.lab,sub.lab)
    }
    return(list(order.df,order.lab))
}
```

```r
test=OrderLab(data=correlation_mat,lab = annlab)
correlation_mat=as.matrix(test[[1]])
annlab=test[[2]]
```
