```r
library(Seurat)
library(ggplot2)
library(NMF)
library(stringr)
source("tools.R")
```

```r
data_dir="./mouse_skin_bulk_cell_RNAseq_weiya_genecount.csv"
geneCounts=read.csv(data_dir,header = TRUE)
genes=as.character(geneCounts$gene)

geneCounts=geneCounts[!duplicated(genes),]
rownames(geneCounts)=as.character(geneCounts$gene)
geneCounts=geneCounts[,-c(1,2,3)]
colnames(geneCounts)<-paste0("P",1:9)
```

#### 去掉Mt, LOC, Gm开头的基因

```r
mt.genes=rownames(geneCounts)[str_detect(string = rownames(geneCounts),pattern = "^mt-")]
Gm.genes=rownames(geneCounts)[str_detect(string = rownames(geneCounts),pattern = "^Gm\\d+")]
loc.genes=rownames(geneCounts)[str_detect(string = rownames(geneCounts),pattern = "^LOC*")]
keep.genes=setdiff(setdiff(setdiff(rownames(geneCounts),mt.genes),Gm.genes),loc.genes)
```

```r
### create Seurat Oject
tol.obj=CreateSeuratObject(raw.data = geneCounts[keep.genes,],project = "Mul.Cell")
#obj@meta.data$sample=paste0("a.",1:nrow(obj@meta.data))

tol.obj=NormalizeData(tol.obj,display.progress =FALSE)
tol.obj=ScaleData(tol.obj,display.progress = FALSE)
tol.obj=FindVariableGenes(tol.obj,do.plot = FALSE,display.progress = FALSE)


#sub.obj=CreateSeuratObject(raw.data = geneCounts[rownames(geneCounts)[rownames(geneCounts)%in%as.character(selected.gene.df$Gene)],],project = "SubMulCell")
sub.obj=CreateSeuratObject(raw.data = geneCounts[rownames(geneCounts)[rownames(geneCounts)%in%sub.genes],],project = "SubMulCell")

sub.obj=NormalizeData(sub.obj,display.progress =FALSE)
sub.obj=ScaleData(sub.obj,display.progress = FALSE)
sub.obj=FindVariableGenes(sub.obj,do.plot = FALSE,display.progress = FALSE)
### cv2 model for gene selection
#v=get.variable.genes(geneCounts,min.cv2 =2,do.plot = FALSE)
#var.genes<-as.character(rownames(v))[v$p.adj<0.01]
```

#### 每个pathway做一个heatmap

```r
PathWayHeatmap=function(Data,PathWayGene,pathway){
    pathway.gene=as.character(PathWayGene$Gene[PathWayGene$Path==pathway])
    aheatmap(as.matrix(Data[pathway.gene[pathway.gene%in%rownames(Data)],]),Colv = NA,Rowv = NA,main = pathway)
}
```

```r
ID.markers<-function(gene.id,All.markers){
    idx=gene.id$gene%in%rownames(All.markers)
    ID=gene.id$id[idx]
    gene=gene.id$gene[idx]
    sub.markers=All.markers[gene,]
    sub.markers$ID=ID
    return(sub.markers)
}
```

```r
path.markers.table=ID.markers(gene.id = path.gene.id,All.markers = All.markers)
```
